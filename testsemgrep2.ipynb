```python
import json
import os

def load_json(file_path):
    """
    Loads a JSON file from disk and returns its contents as a dict.
    Potential security risk: no validation of file_path or JSON schema.
    """
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

def process_data(data):
    """
    Performs simple processing: counts keys and values, and writes output.
    Good: handles missing keys.
    Risk: writes to a static output file without checking if path is safe.
    """
    result = {}
    result['num_keys'] = len(data)
    result['keys'] = list(data.keys())
    # Example of a dangerous pattern: writing to a fixed output file
    output_path = "output/result.json"
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    with open(output_path, 'w') as out_f:
        json.dump(result, out_f, indent=2)
    return result

def main():
    file_path = input("Enter the JSON filename to load: ")
    data = load_json(file_path)
    result = process_data(data)
    print("Processing complete. Result written to output/result.json")
    print("Summary:", result)

if __name__ == "__main__":
    main()
```
